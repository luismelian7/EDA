```python
import pandas as pd
import numpy as np

file_path = './climate_analysis_data.csv'
climate_df = pd.read_csv(file_path)

climate_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos el estilo para las visualizaciones
sns.set(style="whitegrid")

# Análisis Multivariante
# Mapa de Calor de Correlaciones
plt.figure(figsize=(10, 8))
correlation_matrix = climate_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de Calor de Correlaciones", fontsize=14)
plt.show()
```
