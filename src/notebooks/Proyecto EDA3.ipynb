```python
import pandas as pd
import numpy as np

file_path = './climate_analysis_data.csv'
climate_df = pd.read_csv(file_path)

climate_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos el estilo para las visualizaciones
sns.set(style="whitegrid")

# Análisis Bivariante 
# 1. Relación entre Emisiones y Temperatura Anomalía
plt.figure(figsize=(8, 6))
sns.scatterplot(data=climate_df, x='Emisiones Totales (MtCO2e)', y='Temperatura Anomalía (°C)', color='blue')
plt.title("Relación entre Emisiones Totales y Temperatura Anomalía", fontsize=14)
plt.xlabel("Emisiones Totales (MtCO2e)", fontsize=12)
plt.ylabel("Temperatura Anomalía (°C)", fontsize=12)
plt.show()

# 2. Relación entre Temperatura Anomalía y Eventos Climáticos Extremos
plt.figure(figsize=(8, 6))
sns.scatterplot(data=climate_df, x='Temperatura Anomalía (°C)', y='Eventos Climáticos Extremos', color='green')
plt.title("Relación entre Temperatura Anomalía y Eventos Climáticos Extremos", fontsize=14)
plt.xlabel("Temperatura Anomalía (°C)", fontsize=12)
plt.ylabel("Eventos Climáticos Extremos", fontsize=12)
plt.show()
```
