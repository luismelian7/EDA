```python
import pandas as pd
import numpy as np
```

```python
file_path = 'climate_analysis_data.csv'
climate_df = pd.read_csv(file_path)

climate_df.head()

```

```python


# Definimos rangos de años y emisiones por sector
years = list(range(1990, 2025))
sectors = ["Transporte", "Industria", "Energía", "Agricultura", "Residencial"]

# Emisiones totales por año (en millones de toneladas de CO2 equivalente)
total_emissions = np.random.uniform(2500, 5000, len(years)).round(2)

# Dividimos las emisiones por sector (proporcionalmente)
emissions_by_sector = {
    sector: (total_emissions * np.random.uniform(0.1, 0.4)).round(2) for sector in sectors
}

# Temperatura promedio global por año (en °C por encima de la media preindustrial)
temperature_anomalies = np.linspace(0.2, 1.5, len(years)) + np.random.normal(0, 0.1, len(years))

# Número de eventos climáticos extremos por año
extreme_events = np.random.poisson(lam=30, size=len(years)) + np.arange(len(years)) // 5

# Combinamos todo en un DataFrame
data = {
    "Año": years,
    "Emisiones Totales (MtCO2e)": total_emissions,
    "Temperatura Anomalía (°C)": temperature_anomalies.round(2),
    "Eventos Climáticos Extremos": extreme_events,
}
data.update(emissions_by_sector)

climate_df = pd.DataFrame(data)

climate_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos el estilo para las visualizaciones
sns.set(style="whitegrid")

# Análisis Univariante
# 1. Emisiones Totales
plt.figure(figsize=(10, 6))
plt.plot(climate_df['Año'], climate_df['Emisiones Totales (MtCO2e)'], marker='o')
plt.title("Emisiones Totales (MtCO2e) por Año", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Emisiones Totales (MtCO2e)", fontsize=12)
plt.show()

# 2. Temperatura Anomalía
plt.figure(figsize=(10, 6))
plt.plot(climate_df['Año'], climate_df['Temperatura Anomalía (°C)'], marker='o', color='orange')
plt.title("Anomalía de Temperatura Global (°C) por Año", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Temperatura Anomalía (°C)", fontsize=12)
plt.show()

# 3. Eventos Climáticos Extremos
plt.figure(figsize=(10, 6))
plt.plot(climate_df['Año'], climate_df['Eventos Climáticos Extremos'], marker='o', color='red')
plt.title("Frecuencia de Eventos Climáticos Extremos por Año", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Eventos Climáticos Extremos", fontsize=12)
plt.show()
```

```python
# Análisis Bivariante 
# 1. Relación entre Emisiones y Temperatura Anomalía
plt.figure(figsize=(8, 6))
sns.scatterplot(data=climate_df, x='Emisiones Totales (MtCO2e)', y='Temperatura Anomalía (°C)', color='blue')
plt.title("Relación entre Emisiones Totales y Temperatura Anomalía", fontsize=14)
plt.xlabel("Emisiones Totales (MtCO2e)", fontsize=12)
plt.ylabel("Temperatura Anomalía (°C)", fontsize=12)
plt.show()

# 2. Relación entre Temperatura Anomalía y Eventos Climáticos Extremos
plt.figure(figsize=(8, 6))
sns.scatterplot(data=climate_df, x='Temperatura Anomalía (°C)', y='Eventos Climáticos Extremos', color='green')
plt.title("Relación entre Temperatura Anomalía y Eventos Climáticos Extremos", fontsize=14)
plt.xlabel("Temperatura Anomalía (°C)", fontsize=12)
plt.ylabel("Eventos Climáticos Extremos", fontsize=12)
plt.show()
```

```python
# Análisis Multivariante
# Mapa de Calor de Correlaciones
plt.figure(figsize=(10, 8))
correlation_matrix = climate_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de Calor de Correlaciones", fontsize=14)
plt.show()
```
